```python
import utils.problema as p
import algoritmos.q_learning as ql
```

```python
file_mapa1="mapas/Mapa1.txt"
file_mapa2="mapas/Mapa2.txt"
file_mapa3="mapas/Mapa3.txt"
file_mapa4="mapas/Mapa4.txt"

file_mapa = file_mapa4

problema = p.Problema(file_mapa, 0.2)
lectura_mapa = problema.lee_mapa(file_mapa)
estados=problema.estados
transiciones = problema.transiciones
recompensas = problema.recompensas
mapa = lectura_mapa[0]
destino = lectura_mapa[1]
```

```python
##################################### Q-LEARNING #########################################

gamma=0.9
alpha=0.5
epsilon=0.5
max_iteraciones=25000
max_pasos=15

modelo_ql = ql.QLearning(transiciones, recompensas,gamma ,alpha ,epsilon ,max_iteraciones , max_pasos)
modelo_ql.entrenar(destino,mapa)
```

```python
politica_ql = modelo_ql.obtener_politica()
problema.politica_actualizar(politica_ql)
```
